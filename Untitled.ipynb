```python
import numpy as np
import pandas as pd
import tensorflow as tf
import io
import cv2
from sklearn.model_selection import train_test_split
import glob
from keras.models import Sequential
from keras.layers import Conv2D,Conv3D
from keras.layers import MaxPool2D
from keras.layers import Flatten
from keras.layers import Dense,Dropout
import keras
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from skimage import io
```

```python
df=pd.read_csv('training.csv')
```

```python
df.head()
```

```python
df['combine'] = df[['x1', 'x2','y1','y2']].values.tolist()
df.head()
```

```python
train_img_list = glob.glob("images/*")
```

```python
label=[]
data=[]

for imagePath in df['image_name'].values:
    imgn='images\\'+ imagePath
    if imgn in train_img_list :
        image = cv2.imread(imgn)
        imgne=np.array(image)
        #imgnew=imgne.flatten().reshape(1,480*640*3)
        image = cv2.resize(image, (160,120))
        print(image.shape)
        data.append(image)
        label.append(imagePath)
```

```python
d=np.array(data)
```

```python
d=d/255
```

```python
d
```

```python
Y_train=df.iloc[:,1:5]
Y_train['centerx']=(Y_train['x1']+Y_train['x2'])/2
Y_train['centery']=(Y_train['y1']+Y_train['y2'])/2
Y_train['width']=(Y_train['x2']-Y_train['x1'])
Y_train['height']=(Y_train['y2']-Y_train['y1'])
```

```python
Y_train=Y_train.drop(['x1','x2','y1','y2'],axis=1)
```

```python
Y_train.head()
Y=df.iloc[:,:]
```

```python
Y_train.head()
```

```python
(trainX, valX, trainY, valY) = train_test_split(d,Y_train, test_size=0.1, random_state=42)
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3,3), activation='relu', 
                 input_shape=[120,160,3]))
#model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPool2D(pool_size=(3, 3)))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(4, activation='relu'))
model.summary()
```

```python
model.compile(loss=keras.losses.MSE,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])

model.fit(trainX, trainY, batch_size=10,validation_data=(valX, valY) , epochs=150, verbose=1)
```

```python
model_json = model.to_json()
with open("model.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("model.h5")
print("Saved model to disk")
```

```python
from keras.models import model_from_json

json_file = open('model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("model.h5")
print("Loaded model from disk")
 
```
